# 【問題1】100日後の米粒の数

```python
import matplotlib.pyplot as plt
%matplotlib inline

# ある日にもらうコメの数
list_n_grains = [1]
# ある日までにもらったコメの累計数
list_total_grains = [1]
# コメの数を１からスタート
kome = 1

for i in range(99):  # 0からスタートしているのでkome=1を含めて１００回もらう
    kome *= 2
    list_n_grains.append(kome)  # ある日にもらったコメの数を追加
    total = list_total_grains[-1] + kome
    list_total_grains.append(total)  # ある日までにもらったコメの累計を追加

print(kome)
print(list_n_grains)
print(list_total_grains)

plt.title("count of kome")
plt.xlabel("date")
plt.ylabel("number of kome")
plt.tick_params(labelsize=20)
plt.plot(list_n_grains)
plt.show()

plt.title("total of kome")
plt.xlabel("date")
plt.ylabel("number of kome")
plt.tick_params(labelsize=20)
plt.plot(list_total_grains)
plt.show()

# 100日後は633825300114114700748351602688粒もらえる
#100日後の累計は1267650600228229401496703205375粒
```

# 【問題2】100日後以外の米粒の数

```python

def compute_sorori_shinzaemon(day):

    list_n_grains = [1]
    list_total_grains = [1]
    kome = 1

    for i in range(day-1):  # 0からスタートしていてkomeのカウントが1からスタートしているので-1
        kome *= 2
        list_n_grains.append(kome)
        total = list_total_grains[-1] + kome
        list_total_grains.append(total)

    return list_n_grains, list_total_grains


list_n_grains, list_total_grains = compute_sorori_shinzaemon(100)  # ７−10月までの日数

print(compute_sorori_shinzaemon(31+31+30+31))  # 123日

plt.title("count of kome")
plt.xlabel("date")
plt.ylabel("number of kome")
plt.tick_params(labelsize=20)
plt.plot(list_n_grains)
plt.show()

plt.title("count of kome")
plt.xlabel("date")
plt.ylabel("number of kome")
plt.tick_params(labelsize=20)
plt.plot(list_total_grains)
plt.show()


# 回答　　受講期間123日で"10633823966279326983230456482242756607粒"もらえる
```

# 【問題3】何人が何日生活できる米か

```python
def kome_survival(amount_of_grains, number_of_people):

    a_person_requirement = 280/0.02

    # ググったところ、30歳女性の1日の推奨摂取カロリー2000キロカロリーの半分(1000キロカロリー)
    #をお米で取るとする
    # １００グラム３５０キロカロリー計算で一人分が２８０グラムとなる。
    # 1粒０.０２グラムである
    survivaldays = amount_of_grains / \
        (number_of_peoples * a_person_requirement)

    return survivaldays


print(kome_survival(10633823966279326983230456482242756607, 10))

# 例えば、受講生を１０人とする
#答えは7.595588547342376e+31日
```
