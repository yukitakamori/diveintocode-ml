# ダミーデータの作成

## 【問題1】乱数の作成

```python
import numpy as np 
import matplotlib.pyplot as plt
np.random.seed(0)

# 平均
mu = (-3, 0)

# 共分散
cov = [[1.0, 0.8], [0.8, 1.0]]

# 500はデータ数
values1 = np.random.multivariate_normal(mu,cov,500)
x1 = values1[:, 0]
y1 = values1[:, 1]

print(values1.shape)
#print(values1)
```

## 【問題2】散布図による可視化

```python
# 背景を白にする
plt.figure(facecolor = "w")

# 散布図をプロットする
plt.scatter(x1,y1, color = 'b')

#ラベル
plt.xlabel('$x$',size=20)
plt.ylabel('$y$',size=20)

# 軸
plt.grid(True)

# グラフを描画
plt.show

```

## 【問題3】ヒストグラムによる可視化

```python
plt.hist(x1, bins = 500)
plt.xlim (-7, 3)
plt.show
```

```python
plt.hist(y1, bins = 500)
plt.xlim (-7, 3)
plt.show
```

## 【問題4】データの追加

### データの追加

```python
# 平均
mu2 = (0, -3)

# 共分散
cov2 = [[1.0, 0.8], [0.8, 1.0]]

# 500はデータ数
values2 = np.random.multivariate_normal(mu2,cov2,500)
x2 = values2[:, 0]
y2 = values2[:, 1]

print(values2.shape)
#print(values2)
```

### 問題1、4それぞれのデータをひとつの散布図として可視化

```python
# 背景を白にする
plt.figure(facecolor = "w")

# 散布図をプロットする
plt.scatter(x1,y1, color = 'b' , label = 0)
plt.scatter(x2,y2, color = 'r', label = 1)

#ラベル
plt.xlabel('$x$',size=20)
plt.ylabel('$y$',size=20)

# 軸
plt.grid(True)

# 凡例を追加
plt.legend()

# グラフを描画
plt.show
```

## 【問題5】データの結合

```python
# concatenateでやってみる

values3 = np.concatenate([values1, values2])
print(values3.shape)
```

```python
# vstackでやってみる

values4 = np.vstack([values1, values2])
print(values4.shape)
```

## 【問題6】ラベル付け

```python
# 0を５００個、１を５００個作り（１０００、１）の形に成型する
#データと結合する

# ０、１を５００個作る
label0 = np.full(500, 0)
label1 = np.full(500, 1)

#(1000,1)にする
label0 = label0.reshape(500, 1)
label1 = label1.reshape(500,1)

label01 = np.concatenate([label0, label1])

#print(label01.shape)

# データにラベルを結合する
values3_label01 = np.concatenate([values3, label01], 1)
print(values3_label01)
```
