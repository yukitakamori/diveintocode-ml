# 二次元配列と勾配問題

## 【問題1】1次関数

```python
import numpy as np

x_ndarray = np.arange(-50, 50.1, 0.1)
print(len(x_ndarray))

y_ndarray = np.array([])
y_ndarray = (1 / 2)*x_ndarray + 1 # x_ndarrayにまとめて関数の計算をした

    
print(y_ndarray)
```

## 【問題2】配列の結合


```python
x_ndarray = x_ndarray.reshape(1001, 1)
y_ndarray = y_ndarray.reshape(1001, 1)

xy_ndarray = np.concatenate([x_ndarray, y_ndarray],1)

print(xy_ndarray)
```

## 【問題3】勾配を求める

```python
x_gradient = x_ndarray[1:] - x_ndarray[:-1]
y_gradient = y_ndarray[1:] - y_ndarray[:-1]

gradient = y_gradient / x_gradient

print(gradient)
```

## 【問題4】グラフを描く

```python
import matplotlib.pyplot as plt
%matplotlib inline

x_ndarray_1000 = x_ndarray[:1000] #xを一個減らす

# 背景を白にする
plt.figure(facecolor = "w")
# 軸
plt.grid(True)

plt.title("graph")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x_ndarray, y_ndarray, color = 'b', label = "y = 1/2*x + 1")
plt.plot(x_ndarray_1000,gradient, color = "r", label = "gradient")
plt.legend()
plt.show()
```

## 【問題5】Pythonの関数化

```python
# y = x*2
def function1(x_ndarray):
    
    y_ndarray = x_ndarray**2
    
    return y_ndarray
```

```python
# y = 2 * (x**2) + 2**x
def function2(x_ndarray):
    
    y_ndarray = 2 * (x_ndarray**2) + 2**x_ndarray

    return y_ndarray
```

```python
# y = np.sin(x**1/2)
def function3(x_ndarray):
    
    y_ndarray = np.sin(x_ndarray**1/2)
    
    return y_ndarray
```

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    
    # 配列作成
    x_ndarray = np.arange(*x_range)
    y_ndarray = function(x_ndarray)
    
    x_shape = x_ndarray.shape
    
    x_ndarray = x_ndarray.reshape(x_shape[0], 1)
    y_ndarray = y_ndarray.reshape(x_shape[0], 1)
    array_xy = np.concatenate([x_ndarray, y_ndarray],1)
    
    x_gradient = x_ndarray[1:] - x_ndarray[:-1]
    y_gradient = y_ndarray[1:] - y_ndarray[:-1]

    gradient = y_gradient / x_gradient
    
    return array_xy, gradient
```

```python
# y = x*2
array_xy_1, gradient1 = compute_gradient(function1)

# 背景を白にする
plt.figure(facecolor = "w")
# 軸
plt.grid(True)

plt.title("graph")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(array_xy_1[:, 0], array_xy_1[:, 1], color = 'b', label = "y = x*2")
plt.plot(array_xy_1[:1000, 0],gradient1, color = "r", label = "gradient")
plt.legend()
plt.show()
```

```python
# y = 2 * (x**2) + 2**x
array_xy_2, gradient2 = compute_gradient(function2)

# 背景を白にする
plt.figure(facecolor = "w")
# 軸
plt.grid(True)

plt.title("graph")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(array_xy_2[:, 0], array_xy_2[:, 1], color = 'b', label = "y = 2 * (x**2) + 2**x")
plt.plot(array_xy_2[:1000, 0],gradient2, color = "r", label = "gradient")
plt.legend()
plt.show()
```

```python
# y = np.sin(x**1/2)
array_xy_3, gradient3 = compute_gradient(function3, x_range=(0, 50, 0.1))

# 背景を白にする
plt.figure(facecolor = "w")
# 軸
plt.grid(True)

plt.title("graph")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(array_xy_3[:, 0], array_xy_3[:, 1], color = 'b', label = "y = np.sin(x**1/2)")
plt.plot(array_xy_3[:499, 0],gradient3, color = "r", label = "gradient")
plt.legend()
plt.show()
```

## 【問題6】最小値を求める

```python
# y = x*2 の最小値を求める

y_ndarray1 = array_xy_1[:, 1]
y1_min = y_ndarray1.min()
y1_min_idx = y_ndarray1.argmin()
print(y1_min)
print(y1_min_idx)

#499,500番目の勾配
print(gradient1[499])
print(gradient1[500])

```

```python
# y = 2 * (x**2) + 2**x　の最小値を求める

y_ndarray2 = array_xy_2[:, 1]
y2_min = y_ndarray2.min()
y2_min_idx = y_ndarray2.argmin()
print(y2_min)
print(y2_min_idx)

#最小値のyの前後、497,498番目の勾配
print(gradient2[497])
print(gradient2[498])
```

```python
# y = np.sin(x**1/2)　の最小値を求める

y_ndarray3 = array_xy_3[:, 1]
y3_min = y_ndarray3.min()
y3_min_idx = y_ndarray3.argmin()
print(y3_min)
print(y3_min_idx)

#最小値のyの前後、219,220番目の勾配
print(gradient3[219])
print(gradient3[220])
```
